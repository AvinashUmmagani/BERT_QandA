```python
modelname = 'deepset/bert-base-cased-squad2'

model = BertForQuestionAnswering.from_pretrained(modelname)
tokenizer = AutoTokenizer.from_pretrained(modelname)
```

```python
from transformers import pipeline
nlp = pipeline('question-answering', model=model, tokenizer=tokenizer)
```

```python
context = """Can you please enable the SS account N11223344 and let us know the reason for sudden disable.

User account has been disabled at 8 am EET today and the associate is part of TCS windows support team and unable to login to domain services .

Also the associates admin account N11223344 is not working as her normal is disabled.
"""
```

```python
nlp({
    'question': 'What was not working?',
    'context': context
})
```

```python
tokenizer(context)
```
